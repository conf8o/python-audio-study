# 座標と基底と係数

日常的に知っている平面(直交座標系)は、標準基底 $ \{(1, 0), (0, 1)\} $の線形包である。

つまりその平面で私たちがよく表現する任意の点$ (x, y) $の正体は標準基底の線形結合$ x(1, 0) + y(0, 1) $である。

ある基底 $ \{(a, b), (c, d)\} $についての**座標表現**とは、基底の線形結合$ x(a, b) + y(c, d) $の**係数のベクトル** $ (x, y) $のことである。

$ x(a, b) + y(c, d) $の値は標準基底についての座標表現を示している。

例:

基底 $ \{(1, 0), (1, 1)\} $から作られる平面座標は以下のようなy軸を45°斜めに傾けたようなイメージとなる。

この基底についての座標として$(3, 4)$を取ってみた。

この座標を直交座標系、つまり標準基底から作られる平面の座標に置き換えると、

$ 3(1, 0) + 4(1, 1) = (7, 4) $となる。

座標平面を重ねてみると、ピタリと一致する。

## 座標表現の便利なところ

ある位置を係数で表現するというところが肝である。ある基底をベースとしてその線形結合の話を展開するのであれば、係数で語ることができる。


# 波基底と座標とフーリエ変換

任意の実数$ \xi $についての波の関数の集合($= \{ e^{2\pi \mathrm{i} n}\cdot\xi | n \in \mathbb{N} \}$)は互いに線形独立であり、その線形包が関数全体を張ることから関数全体の基底であると言える。ここではこれを波基底と呼ぶ。

つまり、あらゆる関数は波の基底($= \{ e^{2\pi \mathrm{i} x \cdot \xi} | x \in \mathbb{R} \}$)についての"座標"で表現できるということだ。

例: 音の離散フーリエ変換

この例ではあらゆる音を作り出す波の座標をもとめる。

波基底の要素数を20000個として、関数の値域を40000個に離散化する。つまり、音の周波数は20000Hzまでで、サンプリング周波数は40000Hzである。
