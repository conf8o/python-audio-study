# 音声分析

(現状の知識：たぶん間違いありまくり)

音声分析ではアナログな音をデータ分析に適したデータに変換する一連の流れがある。アナログな音をデジタルに変換した段階での音響データ自体はめちゃくちゃ次元の多いベクトル(1sあたり44100要素くらい)なので、それをある一区切り(30ms)あたり十数次元のベクトルに落とす。

0. 音の認知
  * メル尺度と呼ばれる周波数と認知する音の高さの関係を示す尺度がある。人間は高周波ほど音の変化に鈍感らしいので、周波数の対数をもって人間の認知する音の高さを測る(2020-11-26/関数の感覚 参照)。「メル」とつく手続きは大体このメル尺度を考慮する。つまり、周波数の対数を考慮する。

1. A/D変換
  * アナログの音をデジタルに変換する。この物理的な方法について知る優先度は低そう。
  * サンプリング定理に基づいて欲しい音の周波数の2倍をサンプリング周波数とする。そうすると欲しいところまでの音を再現できる。人間が聞こえる高域の限界が20000HzくらいなのでCDは大体2倍の44100Hzがよくサンプリング周波数として使われる。(余談)「ハイレゾ」はもっと高域をとっている。ハイレゾの科学的有用性については諸説あるが、超音波を収録しても無駄とか逆に音質が下がるという説もあれば、音のデータをより高精度で処理したい場合に有用であるという説もある。おもろい。

3. 窓関数
  * 音響データを適当な区間で切り取る。バツバツ切り取るのではなく、ハミング関数でゥオゥオゥと切り取る。(30msくらいの範囲を10msくらい毎)
4. フーリエ変換
  * 窓関数で切り取ったデータをそれぞれフーリエ変換して周波数成分を求める。
  * (3.), (4.)を合わせて短期間フーリエ変換と呼んだりする。
5. メルフィルタバンク
  * ここでようやく次元を削減するフェーズとなる。(4.)で得た各区間の周波数成分はまだサンプリング周波数の半分の要素があるので、そのベクトルの次元を下げるために周波数成分をビン分けする。「メル」と名の付くとおり、メル尺度上で等間隔になるようにする。周波数で等間隔には**しない**。20次元くらいがいいらしい。
