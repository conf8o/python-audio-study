## 高域強調

物理的に高周波は媒質中で減衰しやすいので、高周波が相対的に大きくなるように処理する。

一般的には1オクターブあたり6dB強くする。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from highpassfilter import high_pass_filter

def plot_filtered():
    # x軸は対数めもり
    ax = plt.gca()
    ax.set_xscale('log')
    
    six_filter = high_pass_filter(0.97, 32000)
    frequencies = np.arange(100, 8000)
    
    emphasized = [10 * np.log10(np.abs(six_filter(f)) ** 2) for f in frequencies]
    plt.plot(frequencies, emphasized)
    
plot_filtered()
```
