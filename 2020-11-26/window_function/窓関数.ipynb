# 窓関数

自然に考えると、音響信号は時間についての周期関数とは言えない。周期関数と捉えようとしたとき、周期の始めと終わりが飛ぶ可能性が大きいので全体でフーリエ変換するとやばそう(?)

そこで許容できる範囲の時間で音響信号を切り取ると始めと終わりの飛び具合は小さくなりそう(?)。その切り取るための関数が窓関数。

単純なのは以下のように矩形で切り取る関数(矩形窓)。しかしこれも切り取った信号を周期関数として見たときに周期の始めと終わりが飛ぶ可能性は高い。

```python
# N に対し、0 <= n < 300で区切る矩形窓
w.plot_window(300, w.rectangle_window)
```

```python
# 16Hzの波の300msのところまでを切り取って上記のサンプルと照らし合わせてみる。
w.plot_sample_wave(16, 300)
w.plot_clipped(16, 300, w.rectangle_window)
```

### ハミング窓

そこで周期関数としていい感じに見えるように切り取るための窓関数が作られる。いくつか種類があるが、ここでは**ハミング窓**を使う。

```python
# N に対し、0 <= n < 300で区切るハミング窓
w.plot_window(300, w.hamming_window)
```

```python
# 16Hzの波の300msのところまでをハミング窓で切り取る
w.plot_clipped(16, 300, w.hamming_window)
```

```python
# サンプルと照らし合わせてみる。
w.plot_sample_wave(16, 300)
w.plot_clipped(16, 300, w.hamming_window)
```
