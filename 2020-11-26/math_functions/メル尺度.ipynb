## メル尺度

メル尺度と呼ばれる周波数と認知する音の高さの関係を示す尺度がある。人間は高周波ほど音の変化に鈍感らしいので、周波数の対数をとっている。「メル」とつく手続きは大体このメル尺度を考慮する。つまり、周波数の対数を考慮する。

汎用的に周波数をパラメータとして設定できるようになっているが、よく使われるパラメータとして700Hzや1000Hzがある。

数式書くのはめんどくさいので同階層のmel.pyか下記参照

https://librosa.org/doc/main/generated/librosa.mel_frequencies.html

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from mel import mel_scale 

def plot_mel_scale():
    frequencies = np.arange(20000)
    
    stevens = mel_scale(700)
    fant = mel_scale(1000)
    
    m1 = [*map(stevens, frequencies)]
    m2 = [*map(fant, frequencies)]
    
    plt.plot(m1)
    plt.plot(m2)

plot_mel_scale()
```

```python
# librosaのやつと比較
import librosa as lr

def plot_mel_scale_lib():
    frequencies = np.arange(20000)
    
    # librosa
    plt.plot(lr.hz_to_mel(frequencies, htk=True))
    
    # じゃないほう
    stevens = mel_scale(700)
    m = [*map(stevens, frequencies)]
    plt.plot(m)
    
plot_mel_scale_lib()
```
