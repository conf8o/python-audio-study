## $\mathrm{log}_n(x)$

$x$が大きくなるにつれて値は大きくなるが、大きくなる速度は下がる。

$n$が大きいほど速度が下がりやすくなる。($x \geq 1$)

$0 < x < 1$のときはなんだべこれ

しばしば、$\mathrm{log}_\mathrm{e}(x)$は$\mathrm{log}(x)$や$\mathrm{ln}(x)$と書く。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.grid(which='major',color='gray',linestyle='-')

xs = np.linspace(0.01, 5, 100)

l2 = np.log2(xs)
l10 = np.log10(xs)
ln = np.log(xs)

plt.plot(xs, l2, label="log2")
plt.plot(xs, l10, label="log10")
plt.plot(xs, ln, label="ln")

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=12)
```

## $\mathrm{e}^{-x}$

$x$が大きくなるにつれて値は下がっていくが、下がる速度は落ちていく。

```python
e = np.exp(-xs)
pow2 = np.power(2, -xs)

plt.plot(xs, e, label="exp")
plt.plot(xs, pow2, label="pow 2")

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=12)
```

## これ大事

$a^x = \mathrm{e}^{\mathrm{ln}(a^x)} =\mathrm{e}^{x \cdot \mathrm{ln}(a)}$

累乗の微分で使える。

```python
xs = np.linspace(1, 10, 100)
xs = np.log(xs)
xs = np.exp(xs)

plt.plot(xs)
```
